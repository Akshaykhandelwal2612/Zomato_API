```python
# import libraries
import pandas as pd
import matplotlib .pyplot as plt 
import numpy as np
import csv 
#now we read the dataset and then separate out the restaurants and take out the list of restaurants 
di=dict()
d=pd.read_csv("zomato.csv",encoding='latin-1')
b=d[d["Country Code"]==1]
c=b["Restaurant Name"]
# we iterate over the list and conver them to a dictionary with key as restaurant names and value as nimber of outlets 
for i in c:
    di[i]=di.get(i,0)+1
#now we try to sort out the list and print out the top 15 with number of outlets and print them 
di2= sorted(di.items(), key=lambda item: item[1],reverse=True)

di2=di2[:15]
li1=list()
li2=list()
for i in di2:
    li1.append(i[0])
    li2.append(i[1])
plt.bar(li1,li2,width=0.5)
plt.xticks(rotation=60)
print()
```
