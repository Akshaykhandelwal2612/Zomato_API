```python
import pandas as pd
import matplotlib .pyplot as plt 
import numpy as np
import csv 

d=pd.read_csv("zomato.csv",encoding='latin-1')
b=d[d["Country Code"]==1] 
d2=dict()
for ind in b.index:
    d2[b['Aggregate rating'][ind]]=d2.get(b['Aggregate rating'][ind],0)+b["Votes"][ind]
x=list()
y=list()
for i in d2:
    x.append(i)
    y.append(d2[i])
plt.scatter(x,y)
plt.xlabel("ratings")
plt.ylabel("no of votes")
print()
#on analysing the data we observe that the restaurants having  more number of votes are rated high and lie in range of 3.4 to 4.4 stars on the other hand ones with less votes are rated either too low or too high 
```
