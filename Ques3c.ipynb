```python
#import libraries and then separate out the entire data and separate out the indian data from the dataset 
import pandas as pd
import matplotlib .pyplot as plt 
import numpy as np
import csv 
di=dict()
d=pd.read_csv("zomato.csv",encoding='latin-1')
b=d[d["Country Code"]==1]
#now we will find out the data and apppend  with restaurants and votes and find out the top 10 and print and plot the graph
li=list()
for ind in b.index: 
     li.append([b['Restaurant Name'][ind],b['Votes'][ind]])
li= sorted(li, key=lambda item: item[1],reverse=True)
li=li[:10]
yaxis=list()
xaxis=list()
for i in li:
    xaxis.append(i[0])
    yaxis.append(i[1])
xaxis[-3]=xaxis[-3]+" "
plt.bar(xaxis,yaxis)

plt.xticks(rotation=45)

xaxis,yaxis
print()
```
