```python
#import libraries 
import pandas as pd
import matplotlib .pyplot as plt 
import numpy as np
import csv 

#FIlter out the list and get only the content of the INDIAN countries 
d=pd.read_csv("zomato.csv",encoding='latin-1')
b=d[d["Country Code"]==1] 
d1=dict()
d2=dict()
d3=dict()

#now we will find out the average weighted restaurant rating city wise and number of restaurants they have and along with weighted rading 
for ind in b.index:
    d1[b['City'][ind]]=d1.get(b['City'][ind],0)+1
    d2[b['City'][ind]]=d2.get(b['City'][ind],0)+(b["Aggregate rating"][ind]*b["Votes"][ind])
    d3[b["City"][ind]]=d3.get(b["City"][ind],0)+b["Votes"][ind]
d4=dict()
for i in d3:
    x=round(d2[i]/d3[i],2)
    d4[i]=d4.get(i,0)+x
city=list()
rest=list()
weight=list()
for i in d3:
    city.append(i)
    rest.append(d1[i])
    weight.append(d4[i])
city=np.array(city)
weight=np.array(weight)
rest=np.array(rest)

plt.scatter(city,rest,s=weight)
plt.xticks(rotation=30)
print()
```
