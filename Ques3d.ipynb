```python
#import libraries 
import pandas as pd
import matplotlib .pyplot as plt 
import numpy as np
import csv 
#now we will filter the data only for the USA countries and count the number of cusinies and then their occourance in the list and then plot a pie chart for that 
di=dict()
d=pd.read_csv("zomato.csv",encoding='latin-1')
b=d[d["Country Code"]==216]
b=b["Cuisines"]
b=b.dropna()
dic=dict()
for i in b:
    for j in i.split(", "):
        dic[j]=dic.get(j,0)+1
dic= sorted(dic.items(), key=lambda item: item[1],reverse=True)
li1=list()
li2=list()
dic=dic[:10]
for i in dic:
    li1.append(i[0])
    li2.append(i[1])
plt.pie(li2,labels=li1,autopct="%.1f")
print()
```
